```python
# URL del archivo de datos
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/statlog/german/german.data"

data = pd.read_csv(url, delimiter=' ', header=None)

# Renombrar columnas
column_names = ["status", "duration", "credit_history", "purpose", "amount", "savings", "employment",
                "installment_rate", "personal_status_sex", "other_debtors", "residence_since", "property",
                "age", "other_installment_plans", "housing", "existing_credits", "job", "num_people",
                "telephone", "foreign_worker", "credit"]
data.columns = column_names

data.head()
```

**Ejercicio2 (85-15)**
